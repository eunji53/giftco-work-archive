# PDF 텍스트 추출 진단

상품명에 전화번호·발송처 정보가 섞이는 원인을 찾습니다.

```python
import os, re, sys
sys.path.insert(0, '.')  # pdf_parser.py, config.py import용
from PyPDF2 import PdfReader

from config import PDF_DIR

# ★ 확인할 파일명 입력 (문제가 생긴 파일)
TARGET_FILE = '고려기프트_00284.pdf'  # 예: '고려기프트_20230317.pdf'

# 파일 미지정 시 첫 번째 pdf 사용
if not TARGET_FILE:
    for f in os.listdir(PDF_DIR):
        if f.endswith('.pdf'):
            TARGET_FILE = f
            break

print(f'대상 파일: {TARGET_FILE}')
```

```python
# ── 1. PyPDF2 원문 추출 ──
reader = PdfReader(os.path.join(PDF_DIR, TARGET_FILE))
text = ''.join(page.extract_text() or '' for page in reader.pages)

print('=' * 60)
print('[ 원문 전체 텍스트 ]')
print('=' * 60)
print(text)
```

```python
# ── 2. 품목 블록만 추출 ──
from pdf_parser import extract_item_block, split_item_lines

block = extract_item_block(text)
print('=' * 60)
print('[ extract_item_block 결과 ]')
print('=' * 60)
if block:
    print(block)
else:
    print('None — 블록 추출 실패')
```

```python
# ── 3. 줄 분리 후 각 줄 파싱 결과 ──
from pdf_parser import parse_line

if block:
    lines = split_item_lines(block)
    print(f'총 {len(lines)}줄')
    print('=' * 60)
    for i, line in enumerate(lines):
        result = parse_line(line)
        print(f'\n── 줄 {i+1} ──')
        print(f'  원문: {line}')
        print(f'  파싱: {result}')
```

```python
# ── 4. 원문에서 줄 단위로 보기 (PyPDF2가 어떤 순서로 읽는지 확인) ──
print('[ 원문 줄 번호별 출력 ]')
for i, line in enumerate(text.split('\n'), 1):
    if line.strip():
        print(f'{i:4d}: {line}')
```
